```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data={
    "student":["Rubasri", "Harini", "Madhumitha", "Yuvashree", "Kavish"],
    "Fee paid":[300000, 250000, 200000, 155000, 100000 ]
}

df = pd.DataFrame(data)
colors = ["red", "blue", "yellow", "green", "black"]
sns.barplot(x="student", y="Fee paid", data=df, palette=colors)
plt.title("Fee paid by students")
plt.show()
```

#Types of distribution plots

#**BOX PLOT**

```python
import matplotlib.pyplot as plt
Fees=[300000, 250000, 200000, 155000, 100000 ]
plt.boxplot(Fees)
plt.title("Fee paid by students")
plt.show()
```

#**DENSITY PLOT**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data={
    "student":["Rubasri", "Harini", "Madhumitha", "Yuvashree", "Kavish"],
    "Fee paid":[300000, 250000, 200000, 155000, 100000 ]
}
density=sns.kdeplot(data["Fee paid"], shade=True)

df = pd.DataFrame(data)
colors = ["red", "blue", "yellow", "green", "black"]

plt.title("Fee paid by students")
plt.show()
```

#**HISTOGRAM**

```python
import seaborn as sns
Fees=[300000, 250000, 200000, 155000, 100000 ]
sns.histplot(Fees)
plt.title("Fee paid by students")
plt.show()
```
