#NESTED DICTIONARY

```python
#Create nested dictionary with student data
student = {
    "Ruba": {"Math": 81, "Science": 91, "English": 85},
    "Surya": {"Math": 90, "Science": 91, "English": 87},
    "Tanushree": {"Math": 90, "Science": 87, "English": 85},
    "Laksha": {"Math": 81, "Science": 91, "English": 85},
    "Harini": {"Math": 85, "Science": 90, "English": 82},
}
print(student)
```

# For data manipulation using dictionary convert the dictionary to dataframe

```python
import pandas as pd # pandas are library for data manipulation
data = pd.DataFrame(student) # pass the dictionary to dataframe
data
```

```python
data = pd.DataFrame(student).T # pass the dictionary to dataframe
data
```

# Add analysis to columns

```python
data["Average"]= data.mean(axis=1)
# axis=1 denotes row wise access to the dataframe
data
```

```python
data["Math"]= data.mean(axis=0)
# axis=0 denotes column wise access to the dataframe
data
```

# Data Visualizations

```python
import matplotlib.pyplot as plt
# Bar chart for student average performance
plt.figure(figsize=(8,5))
data["Average"].plot(kind="bar", color= "yellow", edgecolor= "black")
plt.title("Student Average Performance")
plt.xlabel("Student Name")
plt.ylabel("Average Score")
plt.show()
```

# How to visualize subjectwise trend

```python
plt.figure(figsize=(8,5))
#data["Average"].plot(kind="bar", color= "yellow", edgecolor= "black")
for subject in ["Math", "Science", "English"]:
  plt.plot(data[subject], marker='*', label=subject)
plt.title("Student Average Performance")
plt.xlabel("Student Name")
plt.ylabel("Average Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
data[["Math", "Science", "English"]].plot(kind="box")
plt.title("Marks distribution by students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
