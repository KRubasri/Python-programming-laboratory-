```python
import matplotlib.pyplot as plt
data={
    "students":["Rubasri", "Harini", "Madhumitha", "Yuvashree", "Kavish"],
    "Fee paid":[300000, 250000, 200000, 155000, 100000 ]
}
plt.barh(data["students"], data["Fee paid"], color="Grey")
plt.title("Fee paid by students")
plt.xlabel("Fee paid")
plt.ylabel("Students")
plt.show()
```

#Pie chart

```python
import matplotlib.pyplot as plt
students=["Rubasri", "Harini", "Madhumitha", "Yuvashree", "Kavish"]
fees=[300000, 250000, 200000, 155000, 100000 ]
plt.pie(fees, labels=students, autopct="%.2f%%", startangle=90)
plt.title("Fee paid by students")
plt.show()
```

# Interaction with html

```python
import plotly.express as px
import pandas as pd

data ={
    "Student":["Rubasri", "Harini", "Madhumitha", "Yuvashree", "Kavish"],
    "Fee paid":[300000, 250000, 200000, 155000, 100000 ]
}
df = pd.DataFrame(data)

fig = px.bar(df, y="Fee paid", x="Student", title="Fees Paid by Students")
fig.write_html("fees_paid_by_students.html", auto_open=True)
fig.show()
```
