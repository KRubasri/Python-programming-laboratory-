# Exception - Types


```python
print("Demonstration of common python exception\n")
```

```python
print("Zero division error:\n")
number = [1,2,3]
try:
  print(number[2])
  result = 10/0
except:
  print("ZeroDivisionError : you cannot divide by zero.\n")
```

```python
print("Value error:\n")
try:
  num = int("abc")
except ValueError:
  print("ValueError cannot convert string to int\n")
```

```python
print("Type error:\n")
try:
    result="2" + 2
except TypeError:
    print("X TypeError: You tried to add string and integer.\n")
```

```python
print("Index error:\n")
try:
      lst=[1,2,3]
      print(lst[3])
except IndexError:
      print("X IndexError: List index out of range.\n")
```

```python
print("Key error:\n")
try:
            my_dict = {'name': 'Anu'}
            print(my_dict['age'])
except KeyError:
            print("X KeyError: 'age' key not found in the dictionary.\n")
```

# Exception Handling

```python
try:
  num1 = int(input("Enter a number:"))
  num2 = int(input("Enter another number:"))
  result = num1 / num2
  print("Result:", result)
except ZeroDivisionError:
  print("Error: You cannot divide by zero.")
except ValueError:
  print("Error: Please enter numbers only.")
else:
  print("Result:",result)
finally:
  print("Program execution completed.")
```
